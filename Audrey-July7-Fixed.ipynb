```python
import pandas as pd
import matplotlib.pyplot as plt
width = 0.35       # the width of the bars: can also be len(x) sequence
new_dict = [{"postal_code": 22227, "17and18": 100, "20and21": 200},
            {"postal_code": 32227, "17and18": 300, "20and21": 400}]
#df.append(new_row, ignore_index=True)
df = pd.DataFrame(new_dict)

labels = ["17and18", "20and21"]
age1 = df[labels[0]]
age2 = df[labels[1]]

fig, ax = plt.subplots()

ax.bar(labels, age1, width,  label=labels[0])
ax.bar(labels, age2, width,  bottom=age1, label=labels[1])

ax.set_ylabel('Population')
ax.set_title('Population grouped by zip code')
ax.legend()

plt.show()
```
